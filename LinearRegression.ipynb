```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

df = pd.read_csv('Advertising.csv')
df.head()
```

```python
X = df[['TV']]
y = df[['sales']]
plt.scatter(X,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression();
model.fit(X,y);

slope = np.asscalar(model.coef_)
intercept = np.asscalar(model.intercept_)
print("y = " + str(slope) + "X + " + str(intercept))
```

```python
xi=[]
yi=[]
for i in range(300):
    xi.append(i)
    yi.append(slope*i + intercept)

plt.plot(xi, yi, color="green", linewidth=2)
plt.scatter(X,y)
plt.show()
```
