```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs

X, y = make_blobs(n_samples=300, centers=2)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
plt.scatter(X[:,0], y, c=y)
plt.show()
```

```python
X_train = X[:,0]
X_train = X_train.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

linReg = LinearRegression()
linReg.fit(X_train, y)
```

```python
slope = np.asscalar(linReg.coef_)
intcpt = linReg.intercept_
slope, intcpt
```

```python
xi = []
yi = []
for i in range(-2, 15):
    xi.append(i)
    yi.append(i*slope + intcpt)
    
plt.plot(xi, yi, linewidth=2)
plt.scatter(X[:,0], y, c=y)
plt.show()
```

```python
# This is solved by sigmoid function
from math import exp

def sigmoid(x):
    logit = x*slope + intcpt
    p = 1 / (1 + (1 / exp(logit)))
    return p

xi = []
yi = []
for i in range(-15, 35):
    xi.append(i)
    yi.append(sigmoid(i))
    
plt.plot(xi, yi, linewidth=2)
plt.scatter(X[:,0], y, c=y)
plt.show()
```
